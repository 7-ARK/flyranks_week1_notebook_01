## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

'''
Research question

Among pages with sufficient search impressions, which pages have a lower CTR than expected after accounting for their average search position and search volume, and should therefore be prioritized for human review?

Decision being improved

The decision is which pages should enter a limited CTR and engagement review queue first.

Output

The final project should produce a ranked opportunity score for each eligible page. A higher score should indicate stronger evidence that the page is under-capturing clicks relative to comparable pages.

The score should consider factors such as:

impressions;
CTR;
average position;
expected CTR for the page's position tier;
the size of the CTR gap;
sessions and engagement context;
content age, freshness or content type where useful.
Action someone could take

A content or search specialist could review the highest-ranked pages and decide whether to:

improve the page title or meta description;
improve alignment with search intent;
improve snippet-friendly page structure;
strengthen the opening section or on-page content;
investigate weak engagement;
leave the page unchanged and continue monitoring it.

The score itself will not automatically prescribe an edit. It will identify pages that deserve investigation.

Cost of a wrong recommendation

A false positive could waste reviewer time or lead someone to change a page that was already performing appropriately. An unnecessary title or content change could even reduce performance.

A false negative could cause the team to overlook a page where a realistic opportunity to gain additional clicks exists.

For this reason, the system should prioritize explainable recommendations, sufficient data volume and human review rather than treating every low CTR value as a problem.

Why data or machine learning may help

CTR depends strongly on average ranking position and may also vary with visibility, intent, content type, freshness and other page characteristics. A data-driven method can create fairer comparisons than using one universal CTR threshold.

Machine learning may later help detect interactions between these signals, but the project must first establish a transparent position-adjusted baseline. A more complicated model will only be useful if it improves the quality of the ranked review list during proper validation.
'''

```python
from pathlib import Path

import numpy as np
import pandas as pd

# This handles common notebook working-directory locations.
possible_paths = [
    Path("data/raw/content_refresh_anonymized.csv"),
    Path("../../data/raw/content_refresh_anonymized.csv"),
    Path("../data/raw/content_refresh_anonymized.csv"),
]

data_path = next(
    (path for path in possible_paths if path.exists()),
    None,
)

if data_path is None:
    raise FileNotFoundError(
        "Could not find data/raw/content_refresh_anonymized.csv. "
        "Check where the notebook kernel is running."
    )

df = pd.read_csv(data_path)

print(f"Loaded: {data_path.resolve()}")
print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns):,}")

df.head()
```

```python
required_columns = [
    "content_id",
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position",
]

missing_columns = [
    column for column in required_columns
    if column not in df.columns
]

if missing_columns:
    raise ValueError(
        f"Missing required columns: {missing_columns}"
    )

df[required_columns].describe().T
```

```python
analysis_df = df.copy()

numeric_columns = [
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position",
]

for column in numeric_columns:
    analysis_df[column] = pd.to_numeric(
        analysis_df[column],
        errors="coerce",
    )

# Remove duplicate content pages for a page-level analysis.
analysis_df = analysis_df.drop_duplicates(
    subset="content_id"
).copy()

# Convert CTR to a rate between 0 and 1 when it appears to be stored
# as a percentage between 0 and 100.
ctr_99th_percentile = analysis_df["ctr"].quantile(0.99)

if ctr_99th_percentile > 1:
    analysis_df["ctr_rate"] = analysis_df["ctr"] / 100
    ctr_format = "percentage"
else:
    analysis_df["ctr_rate"] = analysis_df["ctr"]
    ctr_format = "decimal rate"

print(f"Detected CTR format: {ctr_format}")
print(f"Unique content pages: {len(analysis_df):,}")
print(
    f"Median CTR: "
    f"{analysis_df['ctr_rate'].median():.2%}"
)
```

```python
eligible = analysis_df.loc[
    (analysis_df["impressions_90d"] >= 500)
    & (analysis_df["avg_position"] > 0)
    & (analysis_df["avg_position"] <= 20)
    & analysis_df["ctr_rate"].notna()
].copy()

print(f"All unique pages: {len(analysis_df):,}")
print(f"Eligible visible pages: {len(eligible):,}")
print(
    "Eligible percentage: "
    f"{len(eligible) / len(analysis_df):.2%}"
)
```

```python
analysis_df = df.copy()

numeric_columns = [
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position",
]

for column in numeric_columns:
    analysis_df[column] = pd.to_numeric(
        analysis_df[column],
        errors="coerce",
    )

# Remove duplicate content pages for a page-level analysis.
analysis_df = analysis_df.drop_duplicates(
    subset="content_id"
).copy()

# Convert CTR to a rate between 0 and 1 when it appears to be stored
# as a percentage between 0 and 100.
ctr_99th_percentile = analysis_df["ctr"].quantile(0.99)

if ctr_99th_percentile > 1:
    analysis_df["ctr_rate"] = analysis_df["ctr"] / 100
    ctr_format = "percentage"
else:
    analysis_df["ctr_rate"] = analysis_df["ctr"]
    ctr_format = "decimal rate"

print(f"Detected CTR format: {ctr_format}")
print(f"Unique content pages: {len(analysis_df):,}")
print(
    f"Median CTR: "
    f"{analysis_df['ctr_rate'].median():.2%}"
)
```

```python
eligible["expected_ctr_for_tier"] = (
    eligible.groupby(
        "position_tier",
        observed=True,
    )["ctr_rate"]
    .transform("median")
)

eligible["ctr_gap"] = (
    eligible["expected_ctr_for_tier"]
    - eligible["ctr_rate"]
)

eligible["below_tier_median"] = (
    eligible["ctr_gap"] > 0
)

underperforming = eligible.loc[
    eligible["below_tier_median"]
].copy()

underperforming["directional_click_gap"] = (
    underperforming["ctr_gap"]
    * underperforming["impressions_90d"]
)

print(
    "Pages below their position-tier median CTR: "
    f"{len(underperforming):,}"
)

print(
    "Percentage of eligible pages below benchmark: "
    f"{len(underperforming) / len(eligible):.2%}"
)

print(
    "Directional click-gap estimate: "
    f"{underperforming['directional_click_gap'].sum():,.0f}"
)
```

```python
candidate_columns = [
    "content_id",
    "impressions_90d",
    "clicks_90d",
    "avg_position",
    "position_tier",
    "ctr_rate",
    "expected_ctr_for_tier",
    "ctr_gap",
    "directional_click_gap",
]

top_candidates = (
    underperforming[candidate_columns]
    .sort_values(
        "directional_click_gap",
        ascending=False,
    )
    .head(10)
    .copy()
)

percentage_columns = [
    "ctr_rate",
    "expected_ctr_for_tier",
    "ctr_gap",
]

for column in percentage_columns:
    top_candidates[column] = (
        top_candidates[column] * 100
    ).round(2)

top_candidates = top_candidates.rename(
    columns={
        "ctr_rate": "ctr_pct",
        "expected_ctr_for_tier": "expected_ctr_pct",
        "ctr_gap": "ctr_gap_percentage_points",
    }
)

top_candidates
```

```python
summary_numbers = {
    "unique_pages": len(analysis_df),
    "eligible_pages": len(eligible),
    "eligible_share": len(eligible) / len(analysis_df),
    "below_benchmark_pages": len(underperforming),
    "below_benchmark_share": (
        len(underperforming) / len(eligible)
    ),
    "directional_click_gap": (
        underperforming["directional_click_gap"].sum()
    ),
}

summary_numbers
```
